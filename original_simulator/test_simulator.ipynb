### Test Script of the simulator simulate.py

Run this to see if the simulator is working as expected. You should see a plot of the number of infected individuals over time, for example.

```python
import numpy as np
from simulator import simulate
import matplotlib.pyplot as plt
```

```python
# Run one replicate with specific parameters
rng = np.random.default_rng(42)
infected, rewires, degrees = simulate(beta=0.3, gamma=0.15, rho=0.7, rng=rng)
print(infected.shape)
print(rewires.shape)
print(degrees.shape)
```

```python
plt.plot(infected)
plt.xlabel("Time")
plt.ylabel("Infected fraction")
plt.title("Simulated epidemic")
plt.show()
```
